**7) ВЫВОДЫ**

*  Из второго и третьего графиков мы видим, что мужчины совершают акты самоубийства примерно **в 5 раз чаще**, чем женщины.
*  Экономические причины особенно влияют на число самоубийст в таких возрастных стратах как: 25 - 34 года, 35 - 54 года и 55 - 74 года.
*  В многих возрастных стратах большинство модулей коэффициента корреляции Пирсона составил больше 0,7 с такими признакими как: ВВП, инфляция, стоимость нефти, прожиточный минимум. Из этого получаем, что количество суицидов сильно зависит от ВВП, инфляции, стоимости нефти, прожиточного минимума.
*   Нам удалось создать линейную регрессию, которая по популяции в 12 возрастно-половых категориях, ВВП, инфляции, стоимости нефти и прожиточному минимуму предсказывает число самоубийств.
*  Посчитав коэффициент детерминации, мы получили, что он равен 0,96, из чего можно сделать вывод, что наша модель **очень точна**.
*  Большой неожиданностью для нас оказались коэффициенты корреляции признаков и самоубийств мальчиков в возрасте 5-14 лет. Мы не смогли найти объективные причины данной тенденции.
